```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
# np.seterr(divide = 'raise')


def compute_y(x):
    x = x.flatten()
    term1 = x ** 2
    # term2 = x ** 2
    # term3 = x ** 1
    term2 = 2*x

    return  1/(term1 + term2 + 0.1)



def generate_array(min_value, steps, total_number):

    array = np.linspace(min_value, min_value + steps * (total_number - 1), total_number)
    return array.reshape(1, -1)  # Reshape to 2D array with shape (1, total_number)

    # return np.linspace(min_value, min_value + steps * (total_number - 1), total_number)

def predicted_fun(x_train, phenotype, length):
    x = x_train
    prediction = eval(phenotype)
    if type(prediction)==float:
        now_pred = [prediction for i in range(length)]
        return now_pred
    return prediction
    


def plot_result(x_train_min, x_train_steps, x_train_count, actual_fun, predicted_fun, phenotype):

    X_train = generate_array(x_train_min, x_train_steps, x_train_count)
    # X_test = generate_array(x_test_min, x_test_steps, x_test_count)


    Y_train = actual_fun(X_train)
    Y_pred = predicted_fun(X_train, phenotype, len(X_train.flatten()))

    
    plt.plot(X_train.flatten(), Y_train, label="actual", color="red")
    plt.plot(X_train.flatten(), Y_pred, label="predicted", color="blue")


# plot_result(0, 1, 50,compute_y, predicted_fun, phenotype = "(-2 / ((-3 - x[0] / 3) / x[0]))")

# plot_result(0, 1, 50,compute_y, predicted_fun, phenotype = "(x[0] - (x[0] / (4 / x[0] + -4 - -5)))")

 
```
